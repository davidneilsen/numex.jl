```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# seaborn dark colors (r, g, b)
sdb='#001c7f'
sdr='#8c0800'
sdg='#12711c'
sdm='#a23582'
#seaborn bright colors (r, g, b)
sbb='#023eff'
sbr='#e8000b'
sbg='#1ac938'
sbc='#00d7ff' # cyan
```

```python
fa = np.genfromtxt(fname="X1.dat")
fb = np.genfromtxt(fname="X3.dat")
fbp1 = np.genfromtxt(fname="X3p1.dat")
fbp4 = np.genfromtxt(fname="X3p4.dat")
```

```python
ta1 = fa[:,0]
ea1 = fa[:,4]
ea2 = fa[:,5]
ea3 = fa[:,6]
tb1 = fb[:,0]
eb1 = fb[:,4]
eb2 = fb[:,5]
eb3 = fb[:,6]
tb1p1 = fbp1[:,0]
eb1p1 = fbp1[:,4]
eb2p1 = fbp1[:,5]
eb3p1 = fbp1[:,6]
tb1p4 = fbp4[:,0]
eb1p4 = fbp4[:,4]
eb2p4 = fbp4[:,5]
eb3p4 = fbp4[:,6]
```

```python
fig, ax = plt.subplots()
ax.semilogy(ta1, np.abs(ea1),color=sbr,label='FD')
ax.semilogy(tb1, np.abs(eb1),color=sbb,label='CFD')
ax.semilogy(tb1p1, np.abs(eb1p1),color=sbg,label='CFD_1proc')
ax.semilogy(tb1p4, np.abs(eb1p4),color=sbc,label='CFD_4proc')
#ax.set_xlim(0.0,4.0)
ax.set_xlabel('t',fontsize='xx-large')
ax.set_ylabel('Error(Ex)',fontsize='xx-large')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig('errEx.pdf')
```

```python
fig, ax = plt.subplots()
ax.semilogy(ta1, np.abs(ea2),color=sbr,label='FD')
ax.semilogy(tb1, np.abs(eb2),color=sbb,label='CFD')
ax.semilogy(tb1p1, np.abs(eb2p1),color=sbg,label='CFD_1proc')
ax.semilogy(tb1p4, np.abs(eb2p4),color=sbc,label='CFD_4proc')
#ax.set_xlim(0.0,4.0)
ax.set_xlabel('t',fontsize='xx-large')
ax.set_ylabel('Error(Ey)',fontsize='xx-large')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig('errEy.pdf')
```

```python
fig, ax = plt.subplots()
ax.semilogy(ta1, np.abs(ea3), color=sbr, label='FD')
ax.semilogy(tb1, np.abs(eb3), color=sbb, label='CFD')
ax.semilogy(tb1p1, np.abs(eb3p1),color=sbg,label='CFD_1proc')
ax.semilogy(tb1p4, np.abs(eb3p4),color=sbc,label='CFD_4proc')
#ax.set_xlim(0.0,4.0)
ax.set_xlabel('t',fontsize='xx-large')
ax.set_ylabel('Error(Bz)',fontsize='xx-large')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig('errBz.pdf')
```
