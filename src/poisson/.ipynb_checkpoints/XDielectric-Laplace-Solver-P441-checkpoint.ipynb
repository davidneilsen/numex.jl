```julia
using Plots
```

```julia
struct Pars
    R0::Int64
    R1::Int64
    R2::Int64
    DL::Int64
    DR::Int64
    DT::Int64
    DB::Int64
    
    dbox::Vector{Float64}
    Vb::Vector{Float64}
    perm1::Float64
    perm2::Float64
    
    function Pars(dielim, V0, perm)
        p1 = 1.0
        new(0, 1, 2, 3, 4, 5, 6, dielim, V0, p1, perm)
    end
end
```

```julia
function initDielectric!(V, perm, mask, x, y, pars)
    nx, ny = size(V)
    V0 = 1.0
    perm1 = pars.perm1
    perm2 = pars.perm2
    
    R0 = pars.R0
    R1 = pars.R1
    R2 = pars.R2
    
    DL = pars.DL
    DR = pars.DR
    DT = pars.DT
    DB = pars.DB
    
    
    Vtop = pars.Vb[1]
    Vbottom = pars.Vb[3]
    Vleft = pars.Vb[4]
    Vright = pars.Vb[2]
    
    dx = x[2] - x[1]
    dy = y[2] - y[1]
    xmin = x[1]
    ymin = y[1]
    
    dxmin = pars.dbox[1]
    dxmax = pars.dbox[2]
    dymin = pars.dbox[3]
    dymax = pars.dbox[4]
    
    dimin = trunc(Int,(dxmin - xmin)/dx) + 1
    dimax = trunc(Int,(dxmax - xmin)/dx) + 1
    djmin = trunc(Int,(dymin - ymin)/dy) + 1
    djmax = trunc(Int,(dymax - ymin)/dy) + 1    
    
    # default values everywhere
    for j = 1:ny
        for i = 1:nx
            V[i,j] = V0
            mask[i,j] = R1
            perm[i,j] = perm1
        end
    end

    # bottom boundary
    for i = 1:nx
        V[i,1] = Vbottom
        mask[i,1] = R0
    end
    
    # top boundary
    for i = 1:nx
        V[i,ny] = Vtop
        mask[i,ny] = R0
    end
    
    # left boundary
    for j = 1:ny
        V[1,j] = Vleft
        mask[1,j] = R0
    end
    
    # right boundary
    for j = 1:ny
        V[nx,j] = Vright
        mask[nx,j] = R0
    end
    
    # dielectric
    for j = djmin+1:djmax-1
        for i = dimin+1:dimax-1
            perm[i,j] = perm2
            mask[i,j] = R2
        end
    end
    
    # dielectric left boundary
    for j = djmin+1:djmax-1
        mask[dimin,j] = DL
    end
    
    #dielectric right boundary
    for j = djmin+1:djmax-1
        mask[dimax,j] = DR
    end
    
    #dielectric top boundary
    for i = dimin+1:dimax-1
        mask[i,djmax] = DT
    end
    
    #dielectric bottom boundary
    for i = dimin+1:dimax-1
        mask[i,djmin] = DB
    end
    
end
```

```julia
# This function defines coordinate arrays.  
# The coordinate x ranges from xmin to xmax with n points.

function defineCoords(n, xmin, xmax)
    x = Array{Float64}(undef,n)
    dx = (xmax - xmin)/(n-1)
    for i = 1:n
        x[i] = xmin + (i-1)*dx
    end
    return x
end
```

```julia
#= This function performs the Gauss-Seidel iteration until either changes in the 
solution drop below a given tolerance, tol, or the maximum number of iterations,
maxSteps is reached.
=#

function GaussSeidel!(V, perm, mask, tol, maxSteps, h, pars)

    nx, ny = size(V)
    step = 0
    
    err = 2.0*tol
    
    while (step < maxSteps && err > tol)
        #print("begin loop, tol=",tol)
        errsum = 0.0
        for j = 2:ny-1
            for i = 2:nx-1
                Vnew = V[i,j]
                if mask[i,j] == pars.DL || mask[i,j] == pars.DR
                    p1 = perm[i+1,j]
                    p2 = perm[i-1,j]
                    Vnew = 0.25*(2.0*p1*V[i+1,j] + 2.0*p2*V[i-1,j] + (p1+p2)*(V[i,j+1]+V[i,j-1]))/(p1+p2)
                elseif mask[i,j] == pars.DT || mask[i,j] == pars.DB
                    p1 = perm[i,j+1]
                    p2 = perm[i,j-1]
                    #@assert abs(p1-p2)<1.0e-6
                    Vnew = 0.25*(2.0*p1*V[i,j+1] + 2.0*p2*V[i,j-1] + (p1+p2)*(V[i-1,j]+V[i+1,j]))/(p1+p2)
                else
                    Vnew = 0.25*(V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1])
                end
                errsum += (Vnew - V[i,j])^2
                V[i,j] = Vnew
            end
        end
        err = errsum*h^2
        step += 1
        #println("Step ",step,", Err ",err)
    end
 
   return step, err
end
```

```julia
#=
#  This routine solves the Laplace equation with Successive Over Relaxation.
#  It also performs the update with a Red-Black ordering of grid points.
#  It is much faster than Gauss-Seidel iteration.
=#
function SOR!(V,omega,perm,mask,tol,maxSteps,h,pars)

    nx, ny = size(V)
    step = 0
    
    err = 2.0*tol
    
    while (step < maxSteps && err > tol)
        #print("begin loop, tol=",tol)
        errsum = 0.0
        for phase = 0:1
            for j = 2:ny-1
                ibgn = 2 + mod(j,2) + phase
                for i = ibgn:2:nx-1
                    Res = 0.0
                    if mask[i,j] == pars.DL || mask[i,j] == pars.DR
                        Res = (perm[i+1,j]*V[i+1,j] + perm[i-1,j]*V[i-1,j])/(perm[i+1,j]+perm[i-1,j]) - V[i,j]
                    elseif mask[i,j] == pars.DT || mask[i,j] == pars.DB
                        Res = (perm[i,j+1]*V[i,j+1] + perm[i,j-1]*V[i,j-1])/(perm[i,j+1]+perm[i,j-1]) - V[i,j]
                    else
                        Res = 0.25*(V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1]- 4.0*V[i,j])
                    end
                    errsum += (omega*Res)^2
                    V[i,j] += omega*Res
                end
            end
        end
        err = errsum*h^2
        step += 1
    end
 
   return step, err
end
```

```julia
function solver(shp, bbox, pars, tol, maxit)
    nx = shp[1]
    ny = shp[2]
    mask = Array{Int64,2}(undef,nx,ny)
    V = Array{Float64,2}(undef,nx,ny)
    perm = Array{Float64,2}(undef,nx,ny)
    x = defineCoords(nx, bbox[1], bbox[2])
    y = defineCoords(ny, bbox[3], bbox[4])
    h = x[2]-x[1]
    initDielectric!(V, perm, mask, x, y, pars)
    #step, err = GaussSeidel!(V, perm, mask, tol, maxit, h, pars)
    omega = 1.5
    step, err = SOR!(V, omega, perm, mask, tol, maxit, h, pars)
    println("Solution finished. step=",step,", err=",err,", tol=",tol)
    return V, x, y, perm, mask, step, err
end
```

```julia
function calE(V, dx, dy)
    nx, ny = size(V)
    Ex = similar(V)
    Ey = similar(V)
    for j = 1:ny
        Ex[1,j] = -(V[2,j]-V[1,j])/dx
        for i = 2:nx-1
            Ex[i,j] = -0.5*(V[i+1,j]-V[i-1,j])/dx
        end
        Ex[nx,j] = -(V[nx,j]-V[nx-1,j])/dx
    end
    
    for i = 1:nx
        Ey[i,1] = -(V[i,2]-V[i,1])/dy
        for j = 2:ny-1
            Ey[i,j] = -0.5*(V[i,j+1]-V[i,j-1])/dy
        end
        Ey[i,ny] = -(V[i,ny]-V[i,ny-1])/dy
    end
    return Ex, Ey
end
```

```julia
dielim = [-0.25, 0.25, -0.25, 0.25]
V0 = [100.0,0.0,0.0,0.0]
p2 = 1.0
p = Pars(dielim,V0,p2)
shp = Int64[101, 101]
bbox = Float64[-1.0,1.0,-1.0,1.0]
tol = 1.0e-6
maxstep = 500000
V, x, y, perm, mask, step, err = solver(shp, bbox, p, tol, maxstep);
```
