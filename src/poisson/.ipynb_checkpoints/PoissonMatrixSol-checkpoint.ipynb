## Solving Gauss' Law

The general form of Gauss' law is
$$ \nabla\cdot{\bf D} = \rho.$$
For a system with linear dielectrics, we can write
$$ {\bf D} = \epsilon{\bf E} = K\epsilon_0{\bf E},$$
where $K$ is the dielectric constant, and $\epsilon_0$ is
the permittivity of vacuum.
The dielectric constant can be a function of position for
an inhomogeneous dielectric, $K = K({\bf r})$.

While Gauss' law can be easily solved analytically for
some simple systems, a numerical solution may be required
for more complicated systems, such as inhomogeneous 
dielectrics.
We will construct a numerical solution here using finite
differences. 

Consider a rectangular domain with Cartesian coordinates
$$x_i = x_{\rm min} + (i-1)\Delta x,\qquad
y_j = y_{\rm min} + (j-1)\Delta y.$$  An index shorthand
will be used for functions of $x$ and $y$, as in
$$V_{ij} = V(x_i,y_j).$$
To derive the finite difference approximation,
we first write
\begin{align}
\nabla\cdot (K{\bf E}) &\approx \frac{1}{\Delta x}\left(K_{i+1/2,j} E^x_{i+1/2,j} - K_{i-1/2,j} E^x_{i-1/2,j}\right)\\
& \qquad + \frac{1}{\Delta y}\left(K_{i,j+1/2} E^y_{i,j+1/2} - K_{i,j-1/2} E^y_{i,j-1/2}\right)
\end{align}
We now write the components of the electric field, $E^x$ and
$E^y$, in terms of the potential
$${\bf E} = - \nabla V.$$
Then
\begin{align}
\nabla\cdot (K{\bf E}) &\approx \frac{1}{\Delta x}\left(K_{i+1/2,j} \left(\frac{V_{i+1,j}-V_{i,j}}{\Delta x}\right) - K_{i-1/2,j}\left(\frac{V_{i,j} - V_{i-1,j}}{\Delta x}\right) \right)\\
& + \frac{1}{\Delta y}\left(K_{i,j+1/2} \left(\frac{V_{i,j+1} - V_{i,j}}{\Delta y}\right) - K_{i,j-1/2} \left(\frac{V_{i,j}-V_{i,j-1}}{\Delta y}\right)\right).
\end{align}
Here the dieletric constant $K$ is evalulated between points
using averages
$$
K_{i\pm 1/2,j} = \frac{1}{2}\left(K_{i\pm 1,j} + K_{ij}\right).
$$
$$\nabla\cdot(K({\bf r}){\bf E}) = \frac{\rho}{\epsilon_0}.$$
We can rewrite the equation as
$$\left(K_{i+1/2,j} V_{i+1,j} + K_{i-1/2,j}V_{i-1,j} + 
K_{i,j+1/2} V_{i,j+1} + K_{i,j-1/2}V_{i,j-1} 
- K_{\rm ave} V_{ij}\right) = -\frac{h^2\rho_{ij}}{\epsilon_0}.
$$
where $K_{\rm ave}$ is the weighted average of the dielectric
constant about the point $(x_i,y_j)$
$$K_{\rm ave} = \frac{1}{2}(K_{i+1,j} + K_{i-1,j} + K_{i,j+1} + K_{i,j-1} + 4K_{ij}).$$

This linear system of equations is solved directly in the form
$$A x = b$$
where $A$ is an $n\times n$ matrix, $x$ is a vector of unknown
potential values, and $b$ is the RHS.

### Notes

1. I don't know the most efficient way to solve the linear system.  Using a $201\times 201$ system, the matrix solution takes about 4 minutes. I haven't been able to solve a $401\times 401$.  I tried using `gmres` in the `InterativeSolvers.jl` package, but it was not successful even for the $201^2$ system.  However, I don't know how to really use the package.

2. The code takes an arbitrary number of dielectrics.  I have written routines to add rectangular or circular regions with constant $K$.

3. Point charges can also be added in the domain.

4. This solution appears to be more accurate than a $401^2$ system solved using Gauss-Seidel iteration with $5\times 10^5$ iterations with SOR.  Iterative methods converge extremely slowly when $K$ is large.

```julia
using Plots
```

```julia
include("utils.jl")
include("potential.jl")
```

```julia
function solver(pars)
    nx = pars.shp[1]
    ny = pars.shp[2]
    mask = Array{Int64,2}(undef,nx,ny)
    perm = Array{Float64,2}(undef,nx,ny)
    x = defineCoords(nx, bbox[1], bbox[2])
    y = defineCoords(ny, bbox[3], bbox[4])
    h = x[2]-x[1]
    Vb = pars.Vb
    initDielectric!(perm, mask, x, y, pars)
    A, b = defA(perm, Vb)
    u = A \ b
    #u = gmres(A, b) # very slow
    V = reshape(u,(nx,ny))
    println("Solution finished.")
    return [V, perm, mask, x, y]
end
```

```julia
shp = [201, 201]
bbox = [-1.0, 1.0, -1.0, 1.0]
Vbound = [100.0,0.0,0.0,0.0]
bb = 0.3
d1 = setSqr([-0.3,-0.3], 0.4, 0.4, 100.0)
d2 = setCyl([0.5,0.5], 0.3, 1000.0)
dielec = [d1, d2]
pars = Pars(shp, bbox, dielec, Vbound)
```

```julia
@time sol = solver(pars)
```

```julia
x = sol[4]
y = sol[5]
dx = x[2]-x[1]
dy = y[2]-y[1]
V = sol[1]
perm = sol[2];
lnperm = similar(perm)
@. lnperm = log10(perm);
```

```julia
contourf(x, y, lnperm',size=(500,500),aspect_ratio=:equal)
```

```julia
contourf(x, y, V',size=(500,500),aspect_ratio=:equal,c=:bluesreds)
```

```julia
Ex = similar(V)
Ey = similar(V)
Exr = similar(V)
Eyr = similar(V)
calE!(Ex, Ey, V, dx, dy);
calE!(Exr, Eyr, V, dx, dy, rescale=true, scale=0.1);
```

```julia
zx = similar(V)
zy = similar(V)
nx,ny = size(V)
for j = 1:ny
    for i = 1:nx
        zx[i,j] = x[i]
        zy[i,j] = y[j]
    end
end
tf=20
zxt = thindata(zx,tf)
zyt = thindata(zy,tf)
Ext = thindata(Ex,tf)
Eyt = thindata(Ey,tf)
Exrt = thindata(Exr,tf)
Eyrt = thindata(Eyr,tf)
En = vectorFieldNorm(Ex, Ey)
@. En = log10(En)
xt = zxt[:,1]
yt = zyt[1,:];
```

```julia
contourf(x, y, En', size=(500,500), aspect_ratio=:equal, c=:bluesreds)
quiver!(zxt, zyt, quiver=(Exrt,Eyrt))
```
